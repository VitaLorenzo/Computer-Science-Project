```python
# rename column to join
movies = movies.rename(columns={'id':'movie_id'})
movies.head()
```

```python
#if movie_id is the unique index, return "false"
def id (df, column):
    if len(df[column]) == len(df[column].drop_duplicates()):
        return "false"
id(credits,'movie_id') and id(movies,'movie_id')
```

```python
# from json column to table
def jsontotable(df, column):
    data = df[column].map(pd.read_json)
    d = pd.DataFrame()
    for i in range(len(df)):
        data[i]['movie_id']=df['movie_id'][i]
        d = d.append(data[i], ignore_index=True)
    return d
```

```python
cast = jsontotable(credits, 'cast')
crew = jsontotable(credits, 'crew')
genres=jsontotable(movies,'genres')
keywords=jsontotable(movies,'keywords')
production_companies=jsontotable(movies,'production_companies')
production_countries=jsontotable(movies,'production_countries')
spoken_languages=jsontotable(movies,'spoken_languages')
```

#### For each company and each decade, compute the overall revenue

```python
movies_year['year'] = movies_year['year'].astype(int)
print(min(movies_year['year']), max(movies_year['year']))
```

```python
def decade (year):
    d = year/10
    return int(d)*10
movies_year['decade'] = movies_year['year'].map(decade)
movies_company = pd.merge(movies_year, production_companies, on='movie_id')
movies_company_decade_revenue = movies_company.groupby(['id_company', 'name', 'decade'])['revenue'].sum().to_frame('Overall revenue')
movies_company_decade_revenue
```

#### For each decade, compute the company with maximum revenue

```python
movies_company_decade_revenue = movies_company_decade_revenue.reset_index(level='decade')
# use set_index to return after the id_company and the name of company (name) through idxmax, that return the index where
# a field is max
movies_company_decade_revenue.groupby('decade')['Overall revenue'].idxmax()
```

#### Distribute the revenue according to the order of appearance in a movie. Assume that the i-th actor contributes twice as much as  the (i+1)-th actor to the revenue.

```python
revenue = movies[['movie_id', 'title', 'revenue']]
rev_actor = pd.merge(cast, revenue, on='movie_id')
rev_actor['order'] = rev_actor['order'].astype(int)
rev_actor.head()
```

```python
# add columns of the value of coefficient's numerator and denominator of each actor of each movie
rev_actor['num_coeff'] = 0.5**(rev_actor['order'])
den = rev_actor.groupby('movie_id')['num_coeff'].sum().to_frame('den_coeff').reset_index()
rev_actor = pd.merge(rev_actor, den, on='movie_id')
rev_actor.head()
```

```python
rev_actor['% of revenue'] = rev_actor['num_coeff']/rev_actor['den_coeff']
rev_actor['Actor revenue'] = rev_actor['% of revenue']*rev_actor['revenue']
rev_actor
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
rev_actor.groupby(['movie_id', 'title']).agg({'revenue':'first', 'Actor revenue':'sum'})
```

#### For each actor find the total revenue attributed to him/her.

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
revenue_actor=rev_actor.groupby(['id', 'name'])['Actor revenue'].sum().to_frame('Total revenue')
revenue_actor
```
